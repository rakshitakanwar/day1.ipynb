Conditional Statement(if, else , elif)


```python
num1=10
if num1>2:
  print("num1: ", num1)
print("if end")
```

```python
num1= 20
num2=10
if num1<=num2:
  print("if part run")
else:
  print("else part run")
```

```python
num1=10
num2=20
num3=30
if num1>num2:
  print("if part run")
elif num1>num3:
  print("elif part run")
else:
  print("else part run")
```

```python
user= int(input("enter the number: "))
num1=20
num2=30
if user>num1:
  print("if part run")
elif user>num2:
  print("elif part run")
else:
  print("else part run")
```

```python
marks=int(input("enter marks: "))
if marks>=95:
  print("A Grade")
elif marks>=90:
  print("B Grade")
elif marks>=75:
  print("C Grade")
elif marks>=60:
  print("D Grade")
else:
  print("FAIL")
```

Dictionary

```python
d= {"name": "Rakshita"}
print(d)
print(d["name"])

#example
d= {"full name": "Rakshita Kanwar"}
print(d["full name"])
```

```python
d= {"name": "rakshita", "id": 300}
print(d["name"])
print(d["id"])
```

```python
d= {"users": ["Ram", "Shyam"]}
print(d['users'][0])
print(d['users'][1])
```

```python
d= {"users":[{"name":"hello"}, {"name": "arya"}]}
print(d['users'][0]["name"])
print(d['users'][1]["name"])
```
