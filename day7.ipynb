```python
class Sakshi:
  def __init(self,bike):
    self.bike=bike

  def usebike(self):
    print("bike use by  :", self.bike)

class Rakshita(Sakshi):
  def __init__(self,name,bike):
    self.bike=bike # first approach
    self.name=name
  def username(self):
    print("name is: ", self.name)

u= Rakshita("rakshita kanwar", "tvs by Rakshita")
u.username()
u.usebike()
```

```python
class Sakshi:
    def __init__(self, bike):
        self.bike = bike

    def usebike(self):
        print("bike use by:", self.bike)


class Rakshita(Sakshi):
    def __init__(self, name, bike):
        self.name = name
        super().__init__(bike)   #second approach

    def username(self):
        print("name is:", self.name)


u = Rakshita("rakshita kanwar", "TVS")
u.username()
u.usebike()
```

```python
#Encapsulation
class Student:
  def __init__(self,name,rollno):
    self.fullname=name   #name , rollno is attribute
    self.rollno=rollno

  def getname(self):
    print(self.fullname)
s= Student("Rakshita", 123)
print(s.fullname)
print(s.rollno)
```

```python
class Student:
  def __init__(self,name,rollno):
    self.__name=name   #name , rollno is attribute
    self.rollno=rollno

  def getname(self):
    print(self.__name)
s= Student("Rakshita", 123)
#print(s.__name) not acess directly
print(s.rollno)
s.getname()  #through this we can access it and __ is used for private or hide the data
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary   # private attribute

    def getsalary(self):
        print(self.__salary)

e = Employee("Rakshita", 50000)
print(e.name)      # accessible
# print(e.__salary)  # not accessible directly
e.getsalary()
```

```python
#overriding
class User1:
  def __init__(self,pen,pencil):
    self.pen = pen
    self.pencil = pencil
  def getpen(self):
    print("user1: ",self.pen)

class User2(User1):
  def __init__(self,pen,pencil,earser):
    self.earser = earser
    super().__init__(pen,pencil)
  def getpen(self):
    print("user2: ",self.pen)

u = User2("blue","black","doms")
u.getpen()
```
