Handle Missing Values


```python
import pandas as pd
import numpy as np
data={
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
import numpy as np
data={
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['Hello','dheeraj','Raj','ram','vinay']
}
df = pd.DataFrame(data)
print(df)
#null values filter| find out
print(df.isnull())
print(df.isnull().sum()) #sum of null
#dropna| drop complete row
df.dropna()  #delete ho jayega vo part jaha null likha hua hai
print(df)
#fillna | jaha bhi null values hai vaha values fill kar do
df.fillna(5)
```

```python
#pie chart
x=["apple","banana","carrot","mango"]
y=[30,50,70,5]
plt.figure(figsize=(4,4))
plt.pie(y,labels=x, autopct="%1.1f%%",colors=["red","yellow","orange","green"], shadow=True)
plt.show()
```
