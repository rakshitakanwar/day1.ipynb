```python
#numpy | list
import numpy as np
import time

start= time.time()
sum1=0
for i in range(1000000):
  sum1+=i*2
end= time.time()
print("for list",sum1, end-start)

#for numpy| array (array work every element | element wise)
start1= time.time()
sum2=0
arr= np.array(1000000)*2
sum2= arr.sum()
endl= time.time()
print("for array",sum2, endl-start1)
```

```python
#zeros
import numpy as np
arr= np.zeros(5)
print(arr)
#2D
arr1= np.zeros((2,3))  #2 rows and 3 column
print(arr1)
```

```python
#ones
import numpy as np
#arr 1d
arr= np.ones(5)
print(arr)

#2d
arr1= np.ones((2,3))
print(arr1)
```

```python
#full
import numpy as np
#1d
arr= np.full((5),7)
print(arr)

#2d
arr1= np.full((2,3),7)
print(arr1)
```

```python
#shape
#1D
arr= np.zeros(6)
print(arr)
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
#2D
arr= np.zeros((3,2))
print(arr)
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

```python
#Reshape
#1D array
import numpy as np
arr= np.arange(6)
print(arr)

#2D array
import numpy as np
arr1= arr.reshape(2,3) # 2 rows and  3column
print(arr1)
```

```python
#flatten
import numpy as np
#2D
arr= np.arange(6)
arr1= arr.reshape(2,3)
print(arr1)

#flatten
print(arr1.flatten())
```

```python
#ravel
import numpy as np
#2D
arr1= np.arange(6).reshape(2,3)
print(arr1)

#ravel
print(arr.ravel())
```

```python
#ravel
import numpy as np
#2D
arr1= np.arange(6).reshape(2,3)
print(arr1)
#flatten  => copy
arr_new= arr1.flatten()
arr_new[0]=100
print(arr1)
```

```python
#ravel => original ke andar hi changes ho jate hai
import numpy as np
#2D
arr1= np.arange(9).reshape(3,3)
print(arr1)
#ravel
arr_new1= arr1.ravel()
arr_new1[0]=100
print(arr1)
```

```python
#Transpose
import numpy as np
arr = np.arange(9).reshape(3,3)
print(arr)

#transpose
arr1= arr.transpose()
print(arr1)
```
