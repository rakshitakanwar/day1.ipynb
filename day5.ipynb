```python
def cal(*a):
  print(eval(f"{a[0]} {a[-1]} {a[1]}"))
cal(10,20,"+")
cal(10,20,"-")
cal(10,20,"*")
cal(10,20,"//")
cal(10,20,"%")
eval("10+30")
```
