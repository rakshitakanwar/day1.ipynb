```python
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
#   loc=>           1-> index of row , name-> column name
print(df.loc[1,"Name"])

#iloc=>       row index-> 0,  column index->0
print(df.iloc[1,1])
```

```python
#get multiple rows
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)

print(df.loc[[0,2]]) #only 0 and 2 index
print(df.loc[[0,1,2]])
#get rows using iloc
print(df.iloc[[1,2,3]])
```

```python
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
#multi rows and columns
print(df.iloc[[0,1],[0,1]]) # starting 2 rows and 2 columns
#slicing
print(df.iloc[0:3])
print(df.loc[0:3])



#loc[row index, column name]
#iloc[row index , column index]
```

```python
#filter
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
#filter one greater
print(df[df['Emp_ID']>104])
#106
print(df[df['Emp_ID']==106])
#30000 to 50000
#print(df[(condition one) & (condition two)])
print(df[(df['Salary']> 30000) & (df['Salary'] < 50000)])
#sorting ascending
print(df.sort_values("Salary"))
print(df.sort_values("Salary", ascending= False))
#two columns
print(df.sort_values(["Salary", "Experience"]))
```
